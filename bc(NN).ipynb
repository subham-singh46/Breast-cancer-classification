```python
import pandas as pd 

import numpy as np

import sklearn.datasets

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

```

```python
#loading data set from sk-learn

BC_data=sklearn.datasets.load_breast_cancer()
```

```python
print(BC_data)
```

```python
#loading data to a dataframe

df= pd.DataFrame(BC_data.data ,columns=BC_data.feature_names)
```

```python
#adding target col form BC_data to the data frame

df['label']=BC_data.target
```

Data is slightly imbalanced

```python
#grouping data based on labels and finding the mean for all the features based on 0 and 1 , 
# i.e Maliganant and benign

df.groupby('label').mean()
```

we can see that the  mean values are more for malignant as compared to benign 

SEPARATING FEATURES AND TARGET

```python



X=df.drop(columns='label',axis=1)
Y=df['label']
```

splitting data for training and testing

```python
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2 ,random_state=2)
```

```python
#checking training shape 

print(X.shape, X_train.shape, X_test.shape)
```

standarized the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()

X_train_std=scaler.fit_transform(X_train)

X_test_std=scaler.transform(X_test)

#x_train and test has been scaled to standarize the data, we dont need to do it for y_tarin,test as because 
#the values are aleady scaled b/w 0 and 1 
```

BUILDING A NEURAL NETWORK ( Using Tensor flow)

```python
import tensorflow as tf 
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model= keras.Sequential([
       keras.layers.Flatten(input_shape=(30,)),
       keras.layers.Dense(20,activation='relu'),
       keras.layers.Dense(2,activation='sigmoid')
])
```

```python
#compiling the neural network

model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training neural network

history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=20)
```

Visualizing Accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model_Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training_data','validation_data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model_Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training_data','validation_data'],loc='lower right')
```

ACCURACY OF MODEL ON TESTING DATA

```python
loss,accuracy=model.evaluate(X_test_std,Y_test)
```

BUILDING THE PREDICTIVE SYSTEM

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')

```
