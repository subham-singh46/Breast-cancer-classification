CHECKING DISTRIBTION OF TARGET VARIABLE

```python
data['diagnosis'].value_counts()
```

label encoding

```python
label_encode= LabelEncoder()

labels=label_encode.fit_transform(data['diagnosis'])

data['target']=labels  #created a new coloumn and passed all the encoded data to it.

data.drop(columns='diagnosis', axis=1, inplace=True)


```

```python
data.head()
```

```python
data['target'].value_counts()
```

GROUPING DATA BASED ON TARGET

```python
data.groupby('target').mean()
```

there is a clear difference between the data of M AND B

Summary From EDA:

1> No missing values

2> All values are continious

3> Mean is slightly more than the median therefore the data is rightly skewed

4> slightly imbalance in the datset(Benign cases are more than Malignant)

5> Mean of most fetures are greater for benign compared to Mlignant(Groupby)


DATA VISUALIZATION

```python
#we will use count plot  because we have two categorical data

sns.countplot(x='target',data=data)
```

DISTRIBUTION PLOT FOR ALL COLOUMNS

```python
for coloumn in data:
    print(coloumn)
```

```python
#distribution plot for all coloumn

for coloumn in data:
    sns.displot(x=coloumn,data=data,)

```

we can see that mostly all the data are positively skewed(RIGHT), WHICH means that the mean of the data is larger than the median

```python
#we can also use distplot to analyze the data too

sns.distplot(x=data.radius_mean)


#this will show us the skewness
```

OUTLIERS DETECTION

```python
for coloumn in data:
    plt.figure()
    data.boxplot([coloumn])
```

Building a co-relation matrix

```python
corelation_matrix= data.corr()
```

```python
#constructing a corelation matrix

plt.figure(figsize=(20,20))

sns.heatmap(corelation_matrix,cbar=True,fmt='.1f',annot=True,cmap='Blues')

plt.savefig('corelation heat map')
```

Inference From EDA And Data Visualization

4> slight imbalance in data set

5> Mean is more for Malignant than Benign

6> Most features have outliers
